# GVH Dynamic 0.1 — Définitions communes des observables

Ce notebook fixe les définitions mathématiques communes des observables GVH utilisées dans les pipelines :

- Pantheon
- H(z)
- BAO
- Lorenz
- Rössler
- Duffing

L’objectif est de garantir que tous les résultats produits par les différents notebooks soient comparables.

Cette version est descriptive, méthodologique et indépendante de toute interprétation physique Volume Partagé.

# 2. Objet géométrique général

Les observables GVH sont définies à partir d'un ensemble discret de points représentant une trajectoire, une série de mesures ou un catalogue de données.

On considère un ensemble :

\[
X_i \in \mathbb{R}^{d},
\qquad
i=1,\ldots,N
\]

où :

- \(N\) est le nombre de points observés ;
- \(d\) est la dimension de l'espace considéré.

Selon le pipeline étudié, les points \(X_i\) peuvent représenter :

- une trajectoire dynamique :
  \[
  X_i=(x_i,y_i,z_i)
  \]
  (Lorenz, Rössler, Duffing) ;

- une série cosmologique :
  \[
  X_i=(z_i,H_i),
  \]
  ou
  \[
  X_i=(z_i,\mu_i),
  \]
  pour les données H(z) ou Pantheon ;

- une série BAO :
  \[
  X_i=(z_i,D_{M,i}/r_d),
  \]
  ou toute autre observable mesurée.

L'idée fondamentale de GVH Dynamic est d'utiliser une définition unique des observables géométriques, indépendamment de la nature physique des données.

Toutes les quantités introduites dans les sections suivantes sont construites uniquement à partir de la géométrie discrète des points \(X_i\).

Ainsi, les pipelines Pantheon, H(z), BAO, Lorenz, Rössler et Duffing reposent exactement sur le même formalisme mathématique.

# 3. Définition de $begin:math:text$D\_T$end:math:text$

La première observable GVH est la distance locale discrète entre deux observations successives.

Soit une suite de points

$begin:math:display$
X\_i \\in \\mathbb\{R\}\^\{d\}\,
\\qquad
i\=1\,\\ldots\,N\.
$end:math:display$

On définit le vecteur de variation locale

$begin:math:display$
\\Delta X\_i
\=
X\_\{i\+1\}\-X\_i\.
$end:math:display$

L'observable fondamentale GVH est alors

$begin:math:display$
D\_T\(i\)
\=
\\\|\\Delta X\_i\\\|\,
$end:math:display$

où $begin:math:text$\\\|\\cdot\\\|$end:math:text$ désigne la norme euclidienne.

De manière explicite,

$begin:math:display$
D\_T\(i\)
\=
\\sqrt\{
\\sum\_\{k\=1\}\^\{d\}
\\left\(
X\_\{i\+1\}\^\{\(k\)\}
\-
X\_i\^\{\(k\)\}
\\right\)\^2
\}\.
$end:math:display$

Dans le cas particulier d'une série scalaire,

$begin:math:display$
x\_i\,
$end:math:display$

la définition devient

$begin:math:display$
D\_T\(i\)
\=
\|x\_\{i\+1\}\-x\_i\|\.
$end:math:display$

Ainsi, $begin:math:text$D\_T$end:math:text$ mesure l'amplitude locale de la variation entre deux observations successives.

Cette définition est indépendante de la nature des données étudiées et constitue la brique élémentaire de toutes les observables GVH.

En particulier :

- pour Lorenz, Rössler et Duffing, $begin:math:text$D\_T$end:math:text$ mesure la distance entre deux états successifs de la trajectoire ;
- pour Pantheon, $begin:math:text$D\_T$end:math:text$ mesure la variation locale des observations cosmologiques ;
- pour H(z), $begin:math:text$D\_T$end:math:text$ mesure la variation discrète de la fonction de Hubble ;
- pour BAO, $begin:math:text$D\_T$end:math:text$ mesure la variation locale des observables baryoniques.

Toutes les observables introduites dans les sections suivantes seront construites à partir de cette quantité fondamentale.

# 4. Définition de $begin:math:text$S\_T$end:math:text$

Après avoir défini la variation locale $begin:math:text$D\_T$end:math:text$, on introduit une mesure globale de sa dispersion.

Soit l'ensemble des valeurs

$begin:math:display$
\\\{D\_T\(i\)\\\}\_\{i\=1\}\^\{N\-1\}\.
$end:math:display$

La moyenne est définie par

$begin:math:display$
\\overline\{D\_T\}
\=
\\frac\{1\}\{N\_D\}
\\sum\_\{i\=1\}\^\{N\_D\}
D\_T\(i\)\,
$end:math:display$

où

$begin:math:display$
N\_D\=N\-1
$end:math:display$

est le nombre total de distances locales.

L'observable GVH $begin:math:text$S\_T$end:math:text$ est définie comme l'écart-type des valeurs de $begin:math:text$D\_T$end:math:text$ :

$begin:math:display$
S\_T
\=
\\sqrt\{
\\frac\{1\}\{N\_D\}
\\sum\_\{i\=1\}\^\{N\_D\}
\\left\(
D\_T\(i\)\-\\overline\{D\_T\}
\\right\)\^2
\}\.
$end:math:display$

Ainsi, $begin:math:text$S\_T$end:math:text$ mesure la dispersion globale des variations locales.

- Une faible valeur de $begin:math:text$S\_T$end:math:text$ indique que les variations sont relativement homogènes.
- Une grande valeur de $begin:math:text$S\_T$end:math:text$ traduit une forte hétérogénéité des variations locales.

Une version normalisée peut également être calculée :

$begin:math:display$
S\_T\^\{\\rm norm\}
\=
\\frac\{S\_T\}\{\\overline\{D\_T\}\}\,
$end:math:display$

mais, dans GVH Dynamic 0.1, la définition de référence reste

$begin:math:display$
\\boxed\{
S\_T\=\\sigma\(D\_T\)
\}
$end:math:display$

afin de conserver une définition unique dans tous les pipelines.

Cette observable est calculée de manière identique pour les jeux de données Pantheon, H(z), BAO ainsi que pour les systèmes dynamiques Lorenz, Rössler et Duffing.

# 5. Définition du seuil critique $begin:math:text$D\_c$end:math:text$

Après avoir caractérisé les variations locales $begin:math:text$D\_T$end:math:text$ et leur dispersion $begin:math:text$S\_T$end:math:text$, on introduit un seuil critique permettant de distinguer les variations ordinaires des variations les plus importantes.

Soit la distribution des valeurs

$begin:math:display$
\\\{D\_T\(i\)\\\}\_\{i\=1\}\^\{N\_D\}\.
$end:math:display$

On définit le seuil critique $begin:math:text$D\_c$end:math:text$ comme un quantile de cette distribution :

$begin:math:display$
D\_c\(q\)\=Q\_q\(D\_T\)\,
$end:math:display$

où $begin:math:text$Q\_q$end:math:text$ désigne le quantile d'ordre $begin:math:text$q$end:math:text$.

Dans GVH Dynamic 0.1, la convention retenue est

$begin:math:display$
\\boxed\{
D\_c\=Q\_\{0\.90\}\(D\_T\)
\}
$end:math:display$

c'est-à-dire le 90ᵉ percentile des valeurs de $begin:math:text$D\_T$end:math:text$.

Ainsi, environ 90 % des variations locales sont inférieures ou égales à $begin:math:text$D\_c$end:math:text$, tandis que les 10 % restantes correspondent aux variations les plus importantes du jeu de données.

Le seuil $begin:math:text$D\_c$end:math:text$ ne dépend d'aucune hypothèse physique. Il est entièrement déterminé par la distribution des données observées.

Cette définition présente plusieurs avantages :

- elle est indépendante des unités de mesure ;
- elle est applicable à tous les jeux de données ;
- elle est robuste vis-à-vis des valeurs extrêmes ;
- elle fournit un seuil objectif permettant d'identifier les régimes de fortes variations.

Le même protocole est utilisé pour tous les pipelines GVH :

- Pantheon ;
- H(z) ;
- BAO ;
- Lorenz ;
- Rössler ;
- Duffing.

Le seuil critique $begin:math:text$D\_c$end:math:text$ constitue la référence utilisée pour définir les observables hiérarchiques introduites dans les sections suivantes, en particulier $begin:math:text$R\_\{\\rm super\}$end:math:text$, $begin:math:text$A\_H$end:math:text$ et $begin:math:text$B\_H$end:math:text$.

# 6. Définition de $begin:math:text$R\_\{\\rm super\}$end:math:text$

Une fois le seuil critique $begin:math:text$D\_c$end:math:text$ défini, il est possible de mesurer la proportion des variations locales qui appartiennent au régime super-critique.

Soit l'ensemble des valeurs

$begin:math:display$
\\\{D\_T\(i\)\\\}\_\{i\=1\}\^\{N\_D\}\,
$end:math:display$

où $begin:math:text$N\_D$end:math:text$ est le nombre total de distances locales.

On définit la fonction indicatrice

$begin:math:display$
\\mathbf\{1\}\_\{D\_T\(i\)\>D\_c\}
\=
\\begin\{cases\}
1\,\& \\text\{si \} D\_T\(i\)\>D\_c\,\\\\
0\,\& \\text\{sinon\.\}
\\end\{cases\}
$end:math:display$

L'observable GVH $begin:math:text$R\_\{\\rm super\}$end:math:text$ est alors définie par

$begin:math:display$
R\_\{\\rm super\}
\=
\\frac\{1\}\{N\_D\}
\\sum\_\{i\=1\}\^\{N\_D\}
\\mathbf\{1\}\_\{D\_T\(i\)\>D\_c\}\.
$end:math:display$

Cette quantité représente la fraction des variations locales dont l'amplitude dépasse le seuil critique.

Par construction,

$begin:math:display$
0 \\leq R\_\{\\rm super\} \\leq 1\.
$end:math:display$

Une faible valeur de $begin:math:text$R\_\{\\rm super\}$end:math:text$ indique que seules quelques variations locales atteignent le régime super-critique.

À l'inverse, une valeur plus élevée traduit une plus grande proportion de variations importantes au sein du jeu de données.

Lorsque le seuil est fixé au quantile

$begin:math:display$
D\_c \= Q\_\{0\.90\}\(D\_T\)\,
$end:math:display$

on s'attend, pour une distribution continue idéale, à obtenir une valeur proche de

$begin:math:display$
R\_\{\\rm super\} \\simeq 0\.10\.
$end:math:display$

En pratique, cette valeur peut différer légèrement en raison :

- de la discrétisation des données ;
- de valeurs identiques au voisinage du seuil ;
- d'un nombre limité d'observations.

La définition de $begin:math:text$R\_\{\\rm super\}$end:math:text$ est identique dans tous les pipelines GVH Dynamic :

- Pantheon ;
- H(z) ;
- BAO ;
- Lorenz ;
- Rössler ;
- Duffing.

Cette observable constitue la première mesure hiérarchique du régime super-critique et servira de base à l'estimation des paramètres $begin:math:text$A\_H$end:math:text$ et $begin:math:text$B\_H$end:math:text$.

# 7. Définition de $begin:math:text$A\_H$end:math:text$

Afin de caractériser la structure hiérarchique des variations super-critiques, on étudie l'évolution de la fraction

$begin:math:display$
R\_\{\\rm super\}
$end:math:display$

lorsque le seuil critique $begin:math:text$D\_c$end:math:text$ varie.

Pour une famille de seuils $begin:math:text$D\_c$end:math:text$, on ajuste la relation empirique

$begin:math:display$
R\_\{\\rm super\}\(D\_c\)
\=
A\_H
\\exp\(\-B\_H D\_c\)\,
$end:math:display$

où $begin:math:text$A\_H$end:math:text$ et $begin:math:text$B\_H$end:math:text$ sont deux paramètres libres.

Le paramètre

$begin:math:display$
\\boxed\{A\_H\}
$end:math:display$

est appelé **amplitude hiérarchique**.

Il correspond à la valeur extrapolée de la fraction super-critique lorsque le seuil tend vers zéro :

$begin:math:display$
R\_\{\\rm super\}\(0\)\=A\_H\.
$end:math:display$

En prenant le logarithme de la relation précédente, on obtient

$begin:math:display$
\\ln\\\!\\left\(R\_\{\\rm super\}\\right\)
\=
\\ln\(A\_H\)
\-
B\_H D\_c\,
$end:math:display$

ce qui permet d'estimer $begin:math:text$A\_H$end:math:text$ et $begin:math:text$B\_H$end:math:text$ par ajustement numérique.

Le paramètre $begin:math:text$A\_H$end:math:text$ ne constitue pas une observable indépendante : il est obtenu à partir de l'ajustement de la loi hiérarchique sur les données.

Dans GVH Dynamic 0.1, $begin:math:text$A\_H$end:math:text$ est utilisé comme indicateur descriptif de l'amplitude globale de la hiérarchie des régimes super-critiques.

La même procédure d'estimation est appliquée à tous les pipelines :

- Pantheon ;
- H(z) ;
- BAO ;
- Lorenz ;
- Rössler ;
- Duffing.

Ainsi, les valeurs de $begin:math:text$A\_H$end:math:text$ obtenues dans les différents jeux de données sont directement comparables, puisqu'elles proviennent de la même définition mathématique et du même protocole d'ajustement.

# 8. Définition de $begin:math:text$B\_H$end:math:text$

Le second paramètre de la loi hiérarchique GVH est le coefficient de décroissance $begin:math:text$B\_H$end:math:text$.

En faisant varier le seuil critique $begin:math:text$D\_c$end:math:text$, la fraction des événements super-critiques est modélisée par

$begin:math:display$
R\_\{\\rm super\}\(D\_c\)
\=
A\_H
\\exp\(\-B\_H D\_c\)\.
$end:math:display$

Le paramètre

$begin:math:display$
\\boxed\{B\_H\}
$end:math:display$

est appelé **coefficient de décroissance hiérarchique**.

Il décrit la vitesse à laquelle la proportion de variations super-critiques diminue lorsque le seuil critique augmente.

En prenant le logarithme de la relation précédente, on obtient

$begin:math:display$
\\ln\\\!\\left\(R\_\{\\rm super\}\\right\)
\=
\\ln\(A\_H\)
\-
B\_H D\_c\.
$end:math:display$

Ainsi, $begin:math:text$B\_H$end:math:text$ correspond à la pente (opposée) de la représentation semi-logarithmique de $begin:math:text$R\_\{\\rm super\}$end:math:text$ en fonction de $begin:math:text$D\_c$end:math:text$.

Dans GVH Dynamic 0.1, les paramètres $begin:math:text$A\_H$end:math:text$ et $begin:math:text$B\_H$end:math:text$ sont estimés simultanément par ajustement non linéaire de la loi exponentielle sur les données.

Par convention,

$begin:math:display$
B\_H \> 0
$end:math:display$

lorsque la fraction des événements super-critiques décroît avec l'augmentation du seuil critique.

Une valeur élevée de $begin:math:text$B\_H$end:math:text$ indique une décroissance rapide de $begin:math:text$R\_\{\\rm super\}$end:math:text$, traduisant une hiérarchie fortement concentrée autour des faibles seuils.

À l'inverse, une valeur plus faible de $begin:math:text$B\_H$end:math:text$ correspond à une décroissance plus lente, indiquant une persistance des régimes super-critiques lorsque le seuil augmente.

Comme pour $begin:math:text$A\_H$end:math:text$, le paramètre $begin:math:text$B\_H$end:math:text$ est obtenu par le même protocole d'ajustement dans tous les pipelines GVH :

- Pantheon ;
- H(z) ;
- BAO ;
- Lorenz ;
- Rössler ;
- Duffing.

Cette définition garantit que les valeurs de $begin:math:text$B\_H$end:math:text$ sont directement comparables entre les différents jeux de données.

# 9. Définition des observables angulaires

En complément des observables de distance, GVH Dynamic introduit un ensemble d'observables angulaires permettant de caractériser l'évolution de la direction des variations locales.

À partir des points

$begin:math:display$
X\_i \\in \\mathbb\{R\}\^\{d\}\,
$end:math:display$

on définit les vecteurs de variation locale

$begin:math:display$
v\_i
\=
X\_\{i\+1\}\-X\_i\.
$end:math:display$

Pour deux vecteurs successifs $begin:math:text$v\_i$end:math:text$ et $begin:math:text$v\_\{i\+1\}$end:math:text$, l'angle local est défini par

$begin:math:display$
\\theta\_i
\=
\\arccos
\\left\(
\\frac\{
v\_i \\cdot v\_\{i\+1\}
\}\{
\\\|v\_i\\\|\\\,\\\|v\_\{i\+1\}\\\|
\}
\\right\)\,
$end:math:display$

où :

- $begin:math:text$v\_i \\cdot v\_\{i\+1\}$end:math:text$ est le produit scalaire ;
- $begin:math:text$\\\|v\_i\\\|$end:math:text$ est la norme euclidienne du vecteur ;
- les angles sont exprimés en degrés.

Les observables angulaires GVH sont alors définies comme suit.

## Angle moyen

L'angle moyen est

$begin:math:display$
\\boxed\{
\\theta\_\{\\rm mean\}
\=
\\frac\{1\}\{N\_\\theta\}
\\sum\_\{i\=1\}\^\{N\_\\theta\}
\\theta\_i
\}
$end:math:display$

où $begin:math:text$N\_\\theta$end:math:text$ est le nombre d'angles calculés.

Cette quantité décrit l'orientation moyenne des variations successives.

---

## Angle maximal

L'angle maximal est

$begin:math:display$
\\boxed\{
\\theta\_\{\\max\}
\=
\\max\(\\theta\_i\)
\}
$end:math:display$

Il correspond à la plus forte variation directionnelle observée dans le jeu de données.

---

## Rapport d'opposition $begin:math:text$R\_\{180\}$end:math:text$

Afin de mesurer la fréquence des changements de direction proches d'une opposition, on définit

$begin:math:display$
R\_\{180\}
\=
\\frac\{1\}\{N\_\\theta\}
\\sum\_\{i\=1\}\^\{N\_\\theta\}
\\mathbf\{1\}\_\{\\theta\_i \\ge 180\^\\circ\-\\varepsilon\}\,
$end:math:display$

où

$begin:math:display$
\\mathbf\{1\}
$end:math:display$

désigne la fonction indicatrice.

Dans GVH Dynamic 0.1, la convention retenue est

$begin:math:display$
\\boxed\{\\varepsilon \= 5\^\\circ\.\}
$end:math:display$

Ainsi,

$begin:math:display$
R\_\{180\}
$end:math:display$

représente la fraction des angles satisfaisant

$begin:math:display$
\\theta\_i \\ge 175\^\\circ\.
$end:math:display$

Cette observable quantifie la fréquence des quasi-oppositions directionnelles au sein des données.

---

Les observables

- $begin:math:text$\\theta\_\{\\rm mean\}$end:math:text$,
- $begin:math:text$\\theta\_\{\\max\}$end:math:text$,
- $begin:math:text$R\_\{180\}$end:math:text$,

sont calculées selon la même procédure dans tous les pipelines GVH Dynamic :

- Pantheon ;
- H(z) ;
- BAO ;
- Lorenz ;
- Rössler ;
- Duffing.

Cette définition garantit que les propriétés angulaires mesurées dans différents jeux de données sont directement comparables.

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit


def as_2d_array(X):
    X = np.asarray(X, dtype=float)
    if X.ndim == 1:
        X = X.reshape(-1, 1)
    return X


def compute_vectors(X):
    X = as_2d_array(X)
    return np.diff(X, axis=0)


def compute_D_T(X):
    V = compute_vectors(X)
    return np.linalg.norm(V, axis=1)


def compute_S_T(D_T, normalized=False):
    D_T = np.asarray(D_T, dtype=float)
    if normalized:
        mean = np.mean(D_T)
        if mean == 0:
            return np.nan
        return np.std(D_T) / mean
    return np.std(D_T)


def compute_D_c(D_T, q=0.90):
    D_T = np.asarray(D_T, dtype=float)
    return np.quantile(D_T, q)


def compute_R_super(D_T, D_c):
    D_T = np.asarray(D_T, dtype=float)
    return np.mean(D_T > D_c)


def exponential_decay(D_c_values, A_H, B_H):
    return A_H * np.exp(-B_H * D_c_values)


def compute_AH_BH(D_T, q_grid=None):
    D_T = np.asarray(D_T, dtype=float)

    if q_grid is None:
        q_grid = np.linspace(0.50, 0.95, 20)

    D_c_values = np.array([compute_D_c(D_T, q=q) for q in q_grid])
    R_values = np.array([compute_R_super(D_T, dc) for dc in D_c_values])

    mask = (R_values > 0) & np.isfinite(D_c_values) & np.isfinite(R_values)
    D_c_fit = D_c_values[mask]
    R_fit = R_values[mask]

    if len(D_c_fit) < 3:
        return np.nan, np.nan, D_c_values, R_values

    try:
        popt, _ = curve_fit(
            exponential_decay,
            D_c_fit,
            R_fit,
            p0=(1.0, 1.0),
            maxfev=10000
        )
        A_H, B_H = popt
    except Exception:
        A_H, B_H = np.nan, np.nan

    return A_H, B_H, D_c_values, R_values


def compute_angles(X):
    V = compute_vectors(X)

    if len(V) < 2:
        return np.array([])

    V1 = V[:-1]
    V2 = V[1:]

    dot = np.sum(V1 * V2, axis=1)
    norm1 = np.linalg.norm(V1, axis=1)
    norm2 = np.linalg.norm(V2, axis=1)

    denom = norm1 * norm2
    valid = denom > 0

    cos_theta = np.full_like(dot, np.nan, dtype=float)
    cos_theta[valid] = dot[valid] / denom[valid]
    cos_theta = np.clip(cos_theta, -1.0, 1.0)

    theta = np.degrees(np.arccos(cos_theta))
    return theta[np.isfinite(theta)]


def compute_theta_observables(theta, epsilon=5.0):
    theta = np.asarray(theta, dtype=float)

    if len(theta) == 0:
        return np.nan, np.nan, np.nan

    theta_mean = np.mean(theta)
    theta_max = np.max(theta)
    R_180 = np.mean(theta >= 180.0 - epsilon)

    return theta_mean, theta_max, R_180


def compute_gvh_observables(X, q_critical=0.90, epsilon=5.0):
    D_T = compute_D_T(X)

    S_T = compute_S_T(D_T, normalized=False)
    S_T_norm = compute_S_T(D_T, normalized=True)

    D_c = compute_D_c(D_T, q=q_critical)
    R_super = compute_R_super(D_T, D_c)

    A_H, B_H, D_c_grid, R_grid = compute_AH_BH(D_T)

    theta = compute_angles(X)
    theta_mean, theta_max, R_180 = compute_theta_observables(
        theta,
        epsilon=epsilon
    )

    summary = pd.DataFrame({
        "Observable": [
            "D_T_mean",
            "S_T",
            "S_T_norm",
            "D_c",
            "R_super",
            "A_H",
            "B_H",
            "theta_mean",
            "theta_max",
            "R_180"
        ],
        "Valeur": [
            np.mean(D_T),
            S_T,
            S_T_norm,
            D_c,
            R_super,
            A_H,
            B_H,
            theta_mean,
            theta_max,
            R_180
        ]
    })

    details = {
        "D_T": D_T,
        "theta": theta,
        "D_c_grid": D_c_grid,
        "R_super_grid": R_grid
    }

    return summary, details
```

```python
# Exemple minimal : trajectoire circulaire 2D

t = np.linspace(0, 10, 500)

X_test = np.column_stack([
    np.sin(t),
    np.cos(t)
])

summary_test, details_test = compute_gvh_observables(X_test)

summary_test
```

Remarque : sur une trajectoire circulaire parfaitement régulière, \(D_T\) est presque constant. L'ajustement de \(A_H\) et \(B_H\) peut donc produire un avertissement numérique. Cet avertissement n'indique pas une erreur du code ; il signale seulement que l'exemple minimal est trop régulier pour contraindre correctement la loi hiérarchique.
